```python
# s = 'aaa' 
# s1 = 'abab'
s = 'baaba'
```

```python
from itertools import permutations
from collections import Counter
```

```python
combos = []
for x in range(2,len(s)+1):
    y = list(permutations(list(s),x))
    combos.append(y)
all_combos = []
for y in combos:
    new_list = Counter(y)
    all_combos.append(new_list)
counts = []    
for x in all_combos:
    for x,v in x.items():
        if v > 1:
            counts.append(v)
```
